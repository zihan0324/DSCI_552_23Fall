### (b) Monte-Carlo Simulation

#### (i) Supervised Learning

```python
def split_dataset(X, y, simulation):
    """
    Randomly split dataset into training set and test set (test size = 0.2)
    Select from positive class and negative class respectively (remain the portion of 2 classes in origin dataset)
    Also standardize attributes for training L1-penalized SVM
    Args:
        X: origin attributes
        y: origin class
        simulation: random state
    Return: X_train, X_test, y_train, y_test
    """
    X_train, X_test, y_train, y_test = train_test_split(test_size=0.2, stratify=y, random_state=simulation)
        
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    
    return X_train, X_test, y_train, y_test


def find_param_range(X, y, start_range, end_range, n_params, threshold):
    '''
    Find the proper range of parameters for cross-validation
    Args:
        start_range: very small parameters to find the start point
        end_range: very large parameters to find the end point
        n_params: desired number of parameters for cross-validation
        threshold: accuracy threshold for detemining start and end point
    Return: np.logspace(start, end, n_params)
    '''
    start = 0
    end = 0
    for s in start_range:
        model = LinearSVC(penalty='l1', dual=False, max_iter=100000, C=math.pow(10, s))
        model.fit(X, y)
        if accuracy_score(y, model.predict(X)) > threshold:
            start = s
            break
    
    for e in end_range:
        model = LinearSVC(penalty='l1', dual=False, max_iter=100000, C=math.pow(10, e))
        model.fit(X, y)
        if accuracy_score(y, model.predict(X)) > threshold:
            end = e
            break
    
    return np.logspace(start, end, n_params)
```

```python
# Find a proper range of parameters of SVM
X_train, X_test, y_train, y_test = split_dataset(X, y, 42)
C_range = find_param_range(X_train, y_train, range(-3, 1), range(6, 2, -1), 10, 0.7)
print("Parameter range for L1-penalized SVM is:", C_range)
```

```python
param_grid = {
    'C': C_range
}

def train_L1_svm(X_train, X_test, y_train, y_test):
    model = LinearSVC(penalty='l1', dual=False, max_iter=100000)
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')
    grid_search.fit(X_train, y_train)

    best_model = grid_search.best_estimator_
    y_train_pred = best_model.predict(X_train)
    y_test_pred = best_model.predict(X_test)
    y_train_df = best_model.decision_function(X_train)
    y_test_df = best_model.decision_function(X_test)

    return best_model, y_train_pred, y_test_pred, y_train_df, y_test_df

def report_pred_results(y_train_actual, y_train_pred, y_test_actual, y_test_pred):
    return {
        'accuracy_train': accuracy_score(y_train_actual, y_train_pred),
        'precision_train': precision_score(y_train_actual, y_train_pred),
        'recall_train': recall_score(y_train_actual, y_train_pred),
        'f1_train': f1_score(y_train_actual, y_train_pred),
        'auc_train': roc_auc_score(y_train_actual, y_train_pred),
        'accuracy_test': accuracy_score(y_test_actual, y_test_pred),
        'precision_test': precision_score(y_test_actual, y_test_pred),
        'recall_test': recall_score(y_test_actual, y_test_pred),
        'f1_test': f1_score(y_test_actual, y_test_pred),
        'auc_test': roc_auc_score(y_test_actual, y_test_pred)
        }

def report_ROC_AUC(y_actual, y_pred, y_prob):
    """
    Report prediction results of binary classification
    Args:
        y_pred: predicted y
        y_actual: actual value of y
        y_prob: predicted probabilities of the positive class
    """

    # Compute confusion matrix
    conf_matrix = confusion_matrix(y_actual, y_pred)
    print("Confusion Matrix:\n", conf_matrix)


    # Compute ROC curve and AUC
    fpr, tpr, _ = roc_curve(y_actual, y_prob)
    roc_auc = auc(fpr, tpr)
    print("Area Under Curve:\n", roc_auc)

    # Plot ROC curve
    plt.figure()
    lw = 2
    plt.plot(fpr, tpr, color='darkorange', lw=lw, label='ROC curve (AUC = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC)')
    plt.legend(loc="lower right")
    plt.show()
```

```python
M = 30
results = []

for s in tqdm(range(M)):
    X_train, X_test, y_train, y_test = split_dataset(X, y, s)
    model, y_train_pred, y_test_pred, y_train_df, y_test_df = train_L1_svm(X_train, X_test, y_train, y_test)
    results.append(report_pred_results(y_train, y_train_pred, y_test, y_test_pred))

# Calculate average metrics over M runs
avg_results = pd.DataFrame(results).mean()
print(f"\nResults of {M} simulations:\n{avg_results}")
```

```python
# Plot the ROC and report the confusion matrix for training and testing in one of the runs
X_train, X_test, y_train, y_test = split_dataset(X, y, simulation=2)
model, y_train_pred, y_test_pred, y_train_df, y_test_df = train_L1_svm(X_train, X_test, y_train, y_test)

print("ROC and confusion matrix for training set:")
report_ROC_AUC(y_train, y_train_pred, y_train_df)
print("\nROC and confusion matrix for testing set:")
report_ROC_AUC(y_test, y_test_pred, y_test_df)
```

#### (ii) Semi-Supervised Learning/ Self-training

##### (A)

```python
def train_semi_L1_svm(X_train, X_test, y_train, y_test, simulation: int):
    # Split training data into labeled and unlabeled
    X_train_labeled, X_train_unlabeled, y_train_labeled, _ = \
      train_test_split(X_train, y_train, test_size=0.5, stratify=y_train, random_state=simulation)

    while len(X_train_unlabeled) > 0:
        # Train L1-penalized SVM on labeled data
        model = LinearSVC(penalty='l1', dual=False, max_iter=100000)
        grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')
        grid_search.fit(X_train_labeled, y_train_labeled)
        best_model = grid_search.best_estimator_

        # Find the unlabeled point farthest from the decision boundary
        distances = best_model.decision_function(X_train_unlabeled)
        farthest_index = np.argmax(np.abs(distances))

        # Add this point to the labeled set
        X_train_labeled = np.append(X_train_labeled, [X_train_unlabeled[farthest_index]], axis=0)
        y_train_labeled = np.append(y_train_labeled, best_model.predict([X_train_unlabeled[farthest_index]]))
        X_train_unlabeled = np.delete(X_train_unlabeled, farthest_index, axis=0)

    y_train_pred = best_model.predict(X_train)
    y_test_pred = best_model.predict(X_test)
    y_train_df = best_model.decision_function(X_train)
    y_test_df = best_model.decision_function(X_test)

    return best_model, y_train_pred, y_test_pred, y_train_df, y_test_df
```

##### (B)

```python
M = 30
semi_results = []

for s in tqdm(range(M)):
    X_train, X_test, y_train, y_test = split_dataset(X, y, s)
    
    best_model, y_train_pred, y_test_pred, y_train_df, y_test_df = train_semi_L1_svm(X_train, X_test,\
                                                                                     y_train, y_test, s)
    
    semi_results.append(report_pred_results(y_train, y_train_pred, y_test, y_test_pred))
    
# Calculate average metrics over M runs
avg_semi_results = pd.DataFrame(semi_results).mean()
print(f"\nResults of {M} simulations of Semi-Supervised Learning:\n{avg_semi_results}")
```

```python
# Plot the ROC and report the confusion matrix for training and testing in one of the runs
X_train, X_test, y_train, y_test = split_dataset(X, y, simulation=2)
model, y_train_pred, y_test_pred, y_train_df, y_test_df = train_semi_L1_svm(X_train, X_test,\
                                                                            y_train, y_test, simulation=2)

print("ROC and confusion matrix for training set:")
report_ROC_AUC(y_train, y_train_pred, y_train_df)
print("\nROC and confusion matrix for testing set:")
report_ROC_AUC(y_test, y_test_pred, y_test_df)
```

#### (iii) Unsupervised Learning

##### (A)

To avoid KMeans from being trapped in a local minimum, **run KMeans multiple times with different random initializations of the centroids**. Among mutiple output, the best output in terms of inertia (sum of squared distances to the nearest cluster center) will be chosen. This increases the chance of finding a more optimal global solution.

```python
def _get_closest_points_to_center(center, X, num_points=30):
    distances = np.linalg.norm(X - center, axis=1)
    return np.argsort(distances)[:num_points]

def _majority_poll(labels):
    return np.bincount(labels).argmax()

def softmax(x):
    e_x = np.exp(x)
    return e_x / e_x.sum(axis=1, keepdims=True)

def cal_kmeans_prob(centers, X):
    # Calculate distances of each point to each cluster center
    distances = np.array([np.linalg.norm(X - center, axis=1) for center in centers]).T
    
    # Apply softmax to get probabilities
    probabilities = softmax(distances)
    
    return probabilities[:, 1] 

def train_kmeans(X_train, X_test, y_train, simulation):
    # Initiate KMeans randomly
    kmeans = KMeans(n_clusters=2, n_init='auto', random_state=simulation).fit(X_train)
    centers = kmeans.cluster_centers_
    
    y_train_pred = np.empty_like(y_train)

    for i, center in enumerate(centers):
        closest_idx = _get_closest_points_to_center(center, X_train)
        cluster_label = _majority_poll(y_train[closest_idx])
        y_train_pred[kmeans.labels_ == i] = cluster_label
        
    y_test_pred = np.array([_majority_poll(y_train[kmeans.labels_ == kmeans.predict([x])[0]]) for x in X_test])
        
    return center, y_train_pred, y_test_pred
```

##### (B)

```python
M = 30
kmeans_results = []

# Split dataset into training and test
indices = np.arange(X.shape[0])
train_indices, test_indices = train_test_split(indices, stratify=y, test_size=0.2, random_state=42)

X_train = X.iloc[train_indices].reset_index(drop=True)
X_test = X.iloc[test_indices].reset_index(drop=True)
y_train = y[train_indices].reset_index(drop=True)
y_test = y[test_indices].reset_index(drop=True)

# Standardize data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
for s in range(M):
    centers, y_train_pred, y_test_pred = train_kmeans(X_train, X_test, y_train, s)
    kmeans_results.append(report_pred_results(y_train, y_train_pred,\
                                              y_test, y_test_pred))
    
# Calculate average metrics over M runs
avg_kmeans_results = pd.DataFrame(kmeans_results).mean()
print(f"\nResults of {M} simulations:\n{avg_kmeans_results}")
```

```python
# Report ROC and confusion matrix of one run
centers, y_train_pred, y_test_pred = train_kmeans(X_train, X_test, y_train, 0)
y_train_prob = cal_kmeans_prob(centers, X_train)
y_test_prob = cal_kmeans_prob(centers, X_test)

print("ROC and confusion matrix for training set:")
report_ROC_AUC(y_train, y_train_pred, y_train_prob)
print("\nROC and confusion matrix for testing set:")
report_ROC_AUC(y_test, y_test_pred, y_test_prob)
```

#### (iv) Spectral Clustering

```python
def train_spectral_clustering(X_train, X_test, y_train, simulation):
    spectral_cluster = SpectralClustering(n_clusters=2,\
                                          affinity='rbf',\
                                          gamma=0.1, random_state=simulation).fit(X)
    train_labels_pred = spectral_cluster.fit_predict(X_train)
    test_labels_pred = spectral_cluster.fit_predict(X_test)
    
    cluster_label_0 = _majority_poll(y_train[train_labels_pred == 0])
    cluster_label_1 = _majority_poll(y_train[train_labels_pred == 1])
    y_train_pred = np.where(train_labels_pred == 0, cluster_label_0, cluster_label_1)
    y_test_pred = np.where(test_labels_pred == 0, cluster_label_0, cluster_label_1)
    
    return y_train_pred, y_test_pred
```

```python
def train_passive_L1_SVM(X, y, simulation):
    test_error_rates = []
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=472, random_state=simulation)
    
    # Standardize data
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
  
    perm = np.random.permutation(len(y_train))
    X_train_perm = X_train[perm]
    y_train_perm = y_train[perm]
    
    for j in range(1, 91):  # For 90 SVMs
        # Select data points
        X_train_subset = X_train_perm[:10 * j]
        y_train_subset = y_train_perm[:10 * j]
        #y_train_subset = y_train_subset.to_numpy()
        test_error_rates.append(L1_SVM_test_error_rate(X_train_subset, X_test, y_train_subset, y_test))
        
    return test_error_rates    
        
```

#### (ii) Active

```python
def _get_cloest_points_to_hyperplane(decision_function, num_points):
    return np.argsort(np.abs(decision_function))[:num_points]

def L1_SVM_test_error_rate_model(X_train, X_test, y_train, y_test):
    model = LinearSVC(penalty='l1', dual=False, max_iter=100000)
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')
    grid_search.fit(X_train, y_train)
    
    best_model = grid_search.best_estimator_
    y_test_pred = best_model.predict(X_test)
    
    return 1 - accuracy_score(y_test, y_test_pred), best_model

def train_active_L1_SVM(X, y, simulation):
    test_error_rates = []
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=472, random_state=simulation)
    
    # Standardize data
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    
    selected_indicies = np.random.choice(X_train.shape[0], size=10, replace=False)
    X_pool = X_train[selected_indices]
    y_pool = y_train[selected_indicies]
    X_left = np.delete(X_train, selected_indicies, axis=0)
    y_left = np.delete(y_train, selected_indicies, axis=0)
    test_error, model = L1_SVM_test_error_rate_model(X_pool, X_test, y_pool, y_test)
    test_error_rates.append(test_error)
    decision_function = model.decision_function(X_left)
    
    while X_left.shape[0] > 0:
        selected_indicies = _get_cloest_points_to_hyperplane(decision_function, 10)
        X_pool = np.append(X_pool, X_left[selected_indicies], axis=0)
        y_pool = np.append(y_pool, y_left[selected_indicies], axis=0)
        X_left = np.delete(X_left, selected_indicies, axis=0)
        y_left = np.delete(y_left, selected_indicies, axis=0)
        test_error, model = L1_SVM_test_error_rate_model(X_pool, X_test, y_pool, y_test)
        test_error_rates.append(test_error)
        if X_left.shape[0] == 0:
            break
        decision_function = model.decision_function(X_left)
        
    return test_error_rates
```

```python
warnings.filterwarnings('ignore', category=UserWarning, module='sklearn.*')
M = 50
active_test_error_results = []

for s in tqdm(range(M)):
    active_test_error_results.append(train_active_L1_SVM(X.values, y.values, s))    
```
