### (b) Train a classifier for each label

#### (i) Research

1. Exact Match Ratio

* It calculates the proportion of samples that have all their labels classified
 correctly.

* Exact Match Ratio = $\frac{1}{N} \sum_{i=1}^{N} I(y_i = \hat{y}_i)$

  * $N$ is th number of samples, $y_i$ is the true set of labels for the
i-th sample, $\hat{y}_i$ is the predicted set of labels, $I$ is the indicator function (1 if true, 0 otherwise).


2. Hamming Score / Loss

* It calculates the fraction of the wrong labels to the total number of labels.

* Hamming Score = $1 - \text{Hamming Loss}$

* Hamming Loss = $\frac{1}{N} \frac{1}{L} \sum_{i=1}^{N} \sum_{l=1}^{L} I(y_{il} \neq \hat{y}_{il})$
  * $N$ is th number of samples, $L$ is the number of labels, $y_{il}$ is the true value of the l-th label of the
i-th sample, $\hat{y}_{il}$ is the predicted value, $I$ is the indicator function (1 if true, 0 otherwise).



#### (ii) Train a SVM for each of the labels

```python
X_train = df_train.drop(columns=['Family', 'Genus', 'Species', 'RecordID'])
X_test = df_test.drop(columns=['Family', 'Genus', 'Species', 'RecordID'])

y1_train = df_train['Family']
y1_test = df_test['Family']

y2_train = df_train['Genus']
y2_test = df_test['Genus']

y3_train = df_train['Species']
y3_test = df_test['Species']

y_train = df_train[['Family', 'Genus', 'Species']]
y_test = df_test[['Family', 'Genus', 'Species']]
```

```python
# Define the parameter grid for cross-validation
param_grid = {
    'estimator__C': [0.1, 1, 10, 100],
    'estimator__gamma': [1, 0.1, 0.01, 0.001]
}

# Function to train SVM for each label
def train_svm(X_train, y_train, X_test, y_test, standardize=False):
    if standardize:
        scaler = StandardScaler()
        X_train = scaler.fit_transform(X_train)
        X_test = scaler.transform(X_test)

    model = OneVsRestClassifier(SVC(kernel='rbf'))
    grid_search = GridSearchCV(model, param_grid, cv=10, scoring='accuracy')
    grid_search.fit(X_train, y_train)

    best_model = grid_search.best_estimator_
    predictions = best_model.predict(X_test)
    accuracy = round(accuracy_score(y_test, predictions), 4)

    return best_model, predictions, accuracy, grid_search.best_params_

# Function to report Exact Match Ratio
def cal_emr(y_actual, y_predict):
    n_instances = len(y_actual)

    exact_match_count = 0
    for actual, pred in zip(y_actual, y_predict):
        if (actual == pred).all():
            exact_match_count += 1

    return round(exact_match_count / n_instances, 4)

# Function to report Hamming Score
def cal_hamming_score(y_actual, y_predict):
    n_labels = len(y_actual) * len(y_actual[0])

    correct_label_count = 0
    for actual, pred in zip(y_actual, y_predict):
        correct_label_count += sum(t == p for t, p in zip(actual, pred))

    return round(correct_label_count / n_labels, 4)
```

```python
# Train SVM for each label
model1, y1_predict, accuracy1, params1 = train_svm(X_train, y1_train, X_test, y1_test)
model2, y2_predict, accuracy2, params2 = train_svm(X_train, y2_train, X_test, y2_test)
model3, y3_predict, accuracy3, params3 = train_svm(X_train, y3_train, X_test, y3_test)

# Train SVM for each label with standardized attributes
model1_std, y1_std_predict, accuracy1_std, params1_std = train_svm(X_train, y1_train, X_test, y1_test, standardize=True)
model2_std, y2_std_predict, accuracy2_std, params2_std = train_svm(X_train, y2_train, X_test, y2_test, standardize=True)
model3_std, y3_std_predict, accuracy3_std, params3_std = train_svm(X_train, y3_train, X_test, y3_test, standardize=True)
```

```python
# Print accuracy for each label
print("Results for raw †attributes:")
print(f"Label 1: Accuracy = {accuracy1}, Best Params = {params1}")
print(f"Label 2: Accuracy = {accuracy2}, Best Params = {params2}")
print(f"Label 3: Accuracy = {accuracy3}, Best Params = {params3}")

print("\nResults for standardized attributes:")
print(f"Label 1: Accuracy = {accuracy1_std}, Best Params = {params1_std}")
print(f"Label 2: Accuracy = {accuracy2_std}, Best Params = {params2_std}")
print(f"Label 3: Accuracy = {accuracy3_std}, Best Params = {params3_std}")
```

```python
# Report Exact Match Ratio
print("Results for raw attributes:")
print(f"Exact Match Ratio = {cal_emr(y_test.values, np.array([y1_predict, y2_predict, y3_predict]).T)}")

print("\nResults for standardized attributes:")
print(f"Exact Match Ratio = {cal_emr(y_test.values, np.array([y1_std_predict, y2_std_predict, y3_std_predict]).T)}")
```

```python
# Report Hamming Score
print("Results for raw attributes:")
print(f"Hamming Score = {cal_hamming_score(y_test.values, np.array([y1_predict, y2_predict, y3_predict]).T)}")

print("\nResults for standardized attributes:")
print(f"Hamming Score = {cal_hamming_score(y_test.values, np.array([y1_std_predict, y2_std_predict, y3_std_predict]).T)}")
```

#### (iii) Repeat 1(b)ii with L1-penalized SVMs

```python
param_grid = {
    'C': np.logspace(-4, 6, 20)
}

def train_L1_svm(X_train, y_train, X_test, y_test):
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    model = LinearSVC(penalty='l1', dual=False, max_iter=100000, multi_class='ovr')
    grid_search = GridSearchCV(model, param_grid, cv=10, scoring='accuracy')
    grid_search.fit(X_train, y_train)

    best_model = grid_search.best_estimator_
    predictions = best_model.predict(X_test)
    accuracy = accuracy_score(y_test, predictions)

    return best_model, predictions, accuracy, grid_search.best_params_
```

```python
# Train SVM wiht L1 penalty for each label
L1_model1, L1_y1_predict, L1_accuracy1, L1_params1 = train_L1_svm(X_train, y1_train, X_test, y1_test)
L1_model2, L1_y2_predict, L1_accuracy2, L1_params2 = train_L1_svm(X_train, y2_train, X_test, y2_test)
L1_model3, L1_y3_predict, L1_accuracy3, L1_params3 = train_L1_svm(X_train, y3_train, X_test, y3_test)
```

```python
# Print results
print(f"Label 1: Accuracy = {L1_accuracy1}, Best Params = {L1_params1}")
print(f"Label 2: Accuracy = {L1_accuracy2}, Best Params = {L1_params2}")
print(f"Label 3: Accuracy = {L1_accuracy3}, Best Params = {L1_params3}")
```

```python
# Report Exact Match Ratio
print(f"Exact Match Ratio: {cal_emr(y_test.values, np.array([L1_y1_predict, L1_y2_predict, L1_y3_predict]).T)}")

# Report Hamming Score
print(f"Hamming Score: {cal_hamming_score(y_test.values, np.array([L1_y1_predict, L1_y2_predict, L1_y3_predict]).T)}")
```

#### (iv) Repeat 1(b)iii by using SMOTE or any other method for imbalance

```python
param_grid = {
    'classifier__C': np.logspace(-4, 6, 20)
}


def train_SMOTE_svm(X_train, y_train, X_test, y_test):
    # Create a pipeline that first applies SMOTE and then trains an SVM
    pipeline = Pipeline([
        ('smote', SMOTE(random_state=42)),
        ('scaler', StandardScaler()),
        ('classifier', LinearSVC(penalty='l1', dual=False, max_iter=100000, multi_class='ovr'))
    ])

    grid_search = GridSearchCV(pipeline, param_grid, cv=10, scoring='accuracy')
    grid_search.fit(X_train, y_train)

    best_model = grid_search.best_estimator_
    predictions = best_model.predict(X_test)
    accuracy = accuracy_score(y_test, predictions)

    return best_model, predictions, accuracy, grid_search.best_params_
```

```python
# Train L1-penalized SVM with SMOTE data for each label
SMOTE_model1, SMOTE_y1_predict, SMOTE_accuracy1, SMOTE_params1 = train_SMOTE_svm(X_train, y1_train, X_test, y1_test)
SMOTE_model2, SMOTE_y2_predict, SMOTE_accuracy2, SMOTE_params2 = train_SMOTE_svm(X_train, y2_train, X_test, y2_test)
SMOTE_model3, SMOTE_y3_predict, SMOTE_accuracy3, SMOTE_params3 = train_SMOTE_svm(X_train, y3_train, X_test, y3_test)
```

```python
# Print results
print(f"Label 1: Accuracy = {SMOTE_accuracy1}, Best Params = {SMOTE_params1}")
print(f"Label 2: Accuracy = {SMOTE_accuracy2}, Best Params = {SMOTE_params2}")
print(f"Label 3: Accuracy = {SMOTE_accuracy3}, Best Params = {SMOTE_params3}")
```

```python
# Report Exact Match Ratio
print(f"Exact Match Ratio: {cal_emr(y_test.values, np.array([SMOTE_y1_predict, SMOTE_y2_predict, SMOTE_y3_predict]).T)}")

# Report Hamming Score
print(f"Hamming Score: {cal_hamming_score(y_test.values, np.array([SMOTE_y1_predict, SMOTE_y2_predict, SMOTE_y3_predict]).T)}")
```

## 2. K-Means Clustering on a Multi-Class and Multi-Label Data Set

### (a) Use k-means clustering

Ref: https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html

```python
# Use CH to choose optimal k, choose from 2, 3,.. , max_k 
def kmeans_clustering(X, max_k, simulation):
    best_k = 2
    best_ch_score = 0

    for k in range(2, max_k + 1):
        kmeans = KMeans(n_clusters=k, n_init='auto', random_state=simulation).fit(X)
        labels = kmeans.labels_
        ch_score = calinski_harabasz_score(X, np.array(labels).T)
        if ch_score > best_ch_score:
            best_ch_score = ch_score
            best_k = k

    # Perform KMeans with the best k
    kmeans = KMeans(n_clusters=best_k, n_init='auto', random_state=simulation).fit(X)
    return best_k, kmeans.labels_
```

```python
X = df_mfcc.drop(columns=['Family', 'Genus', 'Species', 'RecordID'])
y = df_mfcc[['Family', 'Genus', 'Species']]
```

```python
# Try with one simulation
kmeans_clustering(X, 50, 42)
```

### (b) Determine which family is the majority

```python
# Determine majority label for each cluster
def cal_major_label(y, n_clusters, cluser_labels):
    major_labels = []
    for i in range(n_clusters):
        index = np.where(cluser_labels == i)
        cluster_labels = y.iloc[index]
        major_label = cluster_labels.apply(lambda x: Counter(x).most_common(1)[0][0])
        major_labels.append(major_label)
    
    return major_labels
```

```python
best_k, kmeans_labels = kmeans_clustering(X, 50, 42)
print(cal_major_label(y, best_k, kmeans_labels))
```

When random_state=42, the above results shows that best_k=2, majority labels for cluster 0 is {'Leptodactylidae', 'Adenomera', 'AdenomeraHylaedactylus'}, cluster 1 is {'Hylidae', 'Hypsiboas', 'HypsiboasCordobae'} 

### (c) Calculate the average Hamming distance, Hamming score, and Hamming loss

```python
# Calculate Hamming distance
def cal_hamming_distance(y, clusters, major_labels):
    return np.mean([np.sum(row != major_labels[clusters[i]]) for i, row in y.iterrows()]) / 3
```

```python
major_labels = cal_major_label(y, best_k, kmeans_labels)
cal_hamming_distance(y, kmeans_labels, major_labels)
```

```python
cal_hamming_score(y.values, [major_labels[cluster] for cluster in kmeans_labels])
```

```python
# Monte-Carlo simulations
n_simulations = 50
hamming_distances = []
hamming_scores = []
hamming_losses = []


for s in range(n_simulations):
    best_k, kmeans_labels = kmeans_clustering(X, max_k=50, simulation=s)
    major_labels = cal_major_label(y, best_k, kmeans_labels)
    hamming_distances.append(cal_hamming_distance(y, kmeans_labels, major_labels))
    hamming_score = cal_hamming_score(y.values, [major_labels[cluster] for cluster in kmeans_labels])
    hamming_scores.append(hamming_score)
    hamming_loss = 1 - hamming_score
    hamming_losses.append(hamming_loss)
    
# Calculate average and standard deviation of Hamming metrics
average_hamming_distance = np.mean(hamming_distances)
std_dev_hamming_distance = np.std(hamming_distances)
average_hamming_score = np.mean(hamming_scores)
std_dev_hamming_score = np.std(hamming_scores)
average_hamming_loss = np.mean(hamming_losses)
std_dev_hamming_loss = np.std(hamming_losses)

print(f"Average Hamming Distance: {average_hamming_distance} (±{std_dev_hamming_distance})")
print(f"Average Hamming Score: {average_hamming_score} (±{std_dev_hamming_score})")
print(f"Average Hamming Loss: {average_hamming_loss} (±{std_dev_hamming_loss})")
```
