```python
def read_AReM_data_files(dir_str: str, start_row: int, col_number: int) -> pd.DataFrame:
    """
    Read all csv files in the provided directory.
    Args:
        dir_str: a certain direcotry containing data files
        start_row: row number start to read (previous rows will be skipped)
        col_number: number of columns that will be used
    Return: df_data_dict
    """
    df_data_dict = {}
    
    for f in os.listdir(dir_str):
        file_name = dir_str.split('/')[-1] + '_' + f
        file_data = pd.read_csv(os.path.join(dir_str, f),
                                skiprows=range(0, start_row), 
                                usecols=range(0, col_number))
        
        # Tackle None value
        if file_data.isna().any().any():
            file_data_values = pd.read_csv(os.path.join(dir_str, f),
                                          sep=' ',
                                          skiprows=range(0, start_row + 1), 
                                          usecols=range(0, col_number),
                                          header=None).values
            file_data_columns = file_data.columns
            
            file_data = pd.DataFrame(file_data_values, columns=file_data_columns)

            
        df_data_dict[file_name] = file_data
        
    return df_data_dict
```

```python
DATA_PATH = '../data/AReM/'
V_START_ROW = 4 # read from 5th row (row number is 4)
V_COL_NUMBER = 7

# Read data files in each folder
folder_list = [d for d in os.listdir(DATA_PATH) if os.path.isdir(os.path.join(DATA_PATH, d))]

data_dict = {}

for folder in folder_list:
    data_dict.update(read_AReM_data_files(os.path.join(DATA_PATH, folder), V_START_ROW, V_COL_NUMBER))
# df_bending1_dict = read_AReM_data_files('../data/AReM/bending1/', V_START_ROW)
# df_bending2_dict = read_AReM_data_files('../data/AReM/bending2/', V_START_ROW)
# df_cycling_dict = read_AReM_data_files('../data/AReM/cycling/', V_START_ROW)
# df_lying_dict = read_AReM_data_files('../data/AReM/lying/', V_START_ROW)
# df_sitting_dict = read_AReM_data_files('../data/AReM/sitting/', V_START_ROW)
# df_standing_dict = read_AReM_data_files('../data/AReM/standing/', V_START_ROW)
# df_walking_dict = read_AReM_data_files('../data/AReM/walking/', V_START_ROW)
```

#### ii. Extraction

minimum, maximum, mean, median, stan- dard deviation, first quartile, and third quartile

```python
C_MINIMUM = 'min'
C_MAXIMUM = 'max'
C_MEAN = 'mean'
C_MEDIAN = 'median'
C_STANDARD_DEVIATION = 'std'
C_FIRST_QUARTILE = 'first_quartile'
C_THIRD_QUARTILE = 'thrid_quartile'

def cal_stats(time_series_data: pd.Series):
    """
    Calculate 7 statitical features from time series data
    Args:
        time_series_data: the origin time series data
    Return:
    """
    data_description = time_series_data.describe()
    stat = {}
    stat[C_MINIMUM] = data_description['min']
    stat[C_MAXIMUM] = data_description['max']
    stat[C_MEAN] = data_description['mean']
    stat[C_MEDIAN] = data_description['50%']
    stat[C_STANDARD_DEVIATION] = data_description['std']
    stat[C_FIRST_QUARTILE] = data_description['25%']
    stat[C_THIRD_QUARTILE] = data_description['75%']
    
    return stat
```

```python
# Calculate statistical features of 88 instances
V_COLUMN_NAMES = ['avg_rss12', 'var_rss12', 'avg_rss13', 'var_rss13', 'avg_rss23', 'var_rss23']
V_STAT_FEATURE_NAMES = []
C_ACTIVITY = 'activity'
V_TIME_DOMAIN_DATA = []

for key, value in data_dict.items():
    instance = key.split('.')[0]
    
    if len(V_STAT_FEATURE_NAMES) == 0:
        for col in V_COLUMN_NAMES:
            V_STAT_FEATURE_NAMES.append(C_MINIMUM + '_' + col)
            V_STAT_FEATURE_NAMES.append(C_MAXIMUM + '_' + col)
            V_STAT_FEATURE_NAMES.append(C_MEAN + '_' + col)
            V_STAT_FEATURE_NAMES.append(C_MEDIAN + '_' + col)
            V_STAT_FEATURE_NAMES.append(C_STANDARD_DEVIATION + '_' + col)
            V_STAT_FEATURE_NAMES.append(C_FIRST_QUARTILE + '_' + col)
            V_STAT_FEATURE_NAMES.append(C_THIRD_QUARTILE + '_' + col)
            
    feature_data = []
    for col in V_COLUMN_NAMES:
        stats = cal_stats(value[col])
        feature_data += list(stats.values())
    
    feature_data.append(instance)
    V_TIME_DOMAIN_DATA.append(feature_data)
    
V_TIME_DOMIAN_COLUMNS = [*V_STAT_FEATURE_NAMES, C_ACTIVITY]
df_time_domain_features = pd.DataFrame(V_TIME_DOMAIN_DATA, columns=V_TIME_DOMIAN_COLUMNS)

print(df_time_domain_features[V_STAT_FEATURE_NAMES].head())
```

#### iii. Standard Deviation

##### Estimate Standard Deviation

```python
C_FEATURE_NAME = 'feature_name'
C_STANDARD_DEVIATION = 'stand_deviation'
feature_std_value = []

for col in V_STAT_FEATURE_NAMES:
    feature_std_value.append(df_time_domain_features[col].std())

df_feature_std = pd.DataFrame({C_FEATURE_NAME: V_STAT_FEATURE_NAMES, C_STANDARD_DEVIATION: feature_std_value})
print(df_feature_std)
```

##### Build  90% bootsrap confidence interval

Ref: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.bootstrap.html

```python
V_BOOTSTRAP_STATISTIC = np.std
V_CONFIDENCE_LEVEL = 0.9
V_RANDOM_STATE = 42
C_LOWER_BOUND = 'lower_bound'
C_HIGHER_BOUND = 'higher_bound'

feature_std_lower_bound = []
feature_std_higher_bound = []

for col in V_STAT_FEATURE_NAMES:
    bs_res = bootstrap((df_time_domain_features[col].values,), V_BOOTSTRAP_STATISTIC,
                       confidence_level=V_CONFIDENCE_LEVEL, random_state=V_RANDOM_STATE)
    
    feature_std_lower_bound.append(bs_res.confidence_interval.low)
    feature_std_higher_bound.append(bs_res.confidence_interval.high)
    
df_feature_std = pd.concat([df_feature_std, pd.DataFrame({C_LOWER_BOUND: feature_std_lower_bound,
                                        C_HIGHER_BOUND: feature_std_higher_bound})], axis=1)

print(df_feature_std)
```

The *lower_bound* and *higher_bound* of min_var_rss12, min_var_rss13 are NaN because these 2 features are always equal to 0 in this sample dataset.

## 2. ISLR 3.7.4

### (a) Linear Train

When the true relationship between X and Y is linear

• **Linear Regression**: This model fits well with the relationship between X and Y, outputs a certain training RSS

• **Cubic Regression**: Though the true relationship between X and Y is linear, there will be some random fluctuations (noise) in the data. The quadratic term ($X^2$) and cubic term ($X^3$) will fit better to these fluctuations, including noise as model pattern. $\beta_2$ and $\beta_3$ should be ideally close to zero, but due to noise in data, they would result in a training RSS at least as well as linear regression.

• **Conclusion**: training RSS of cubic regression ≤ training RSS of linear regression

### (b) Linear Test

When the true relationship between X and Y ia linear

• **Linear Regression**: Given that the true relationship between X and Y is linear, linear regression should fit well with test data. Thus, test RSS of linear regression should be relatively low.

• **Cubic Regression**: Though cubic regression has a lower training RSS than linear regression, it may overfit to the noise in the training data. When cubic regression is applied to unknown test data, it tends to have a higher RSS because its pattern learned from training data won't be same as test data.

• **Conclusion**: test RSS of cubic regression > test RSS of linear regression

### (c) Not Linear Train

When the true relationship between X and Y is not linear

• **Linear Regression**: When the true relationship between X and Y is not linear, linear regression will have limitations to fit the non-linearity in data, which results in a higher training RSS. 

• **Cubic Regression**: Cubic regression can better fit the non-linearity (curve) in data even the true relationship might not be exactly cubic. The quadratic term ($X^2$) and cubic term ($X^3$) provide more flxibility to capture the potential relationship between X and Y. Thus, cubic regression would have a lower training RSS.

• **Conclusion**: training RSS of cubic regression < training RSS of linear regression

### (d) Not Linear Testing

When the true relationship between X and Y is linear

• **Linear Regression**: Linear regression tends to underfit the non-linear relationship between X and Y, so its test RSS would be relatively high.

• **Cubic Regression**: Cubic regression may have better performane in fitting the non-linear relationship between X and Y. 

However, it is also possible that cubic regression overfits the training data, which results in a higher test RSS than linear regression.

• **Conclusion**: there is not enough information to tell
